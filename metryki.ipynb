### Metryki

#### Definicja

Metryką na zbiorze $X$ nazywamy funkcję $d: X \times X \rightarrow [0, \infty)$ spełniająca poniższe warunki.

$$ \forall (x,y,z \in X) $$
1. $ d(x, y) \geq 0$  (funkcja nieujemna)
2. $ d(x, y) = 0 \Leftrightarrow x = y $
3. $d(x, y) = d(y,x)$  (funkcja symetryczna)
4. $d(x, y) \leq d(x, z) + d(z, y)$ (nierówność trójkąta)

### Metryka euklidesowa

Dla punktów $p$ i $q$ takich, że:
$$p \in R^{n} \:\:\:\:\:\:\:\ p=(p_{1}, p_{2}, ..., p_{n})$$
$$q \in R^{n} \:\:\:\:\:\:\:\ q=(q_{1}, q_{2}, ..., q_{n})$$

Metryka euklidesowa ma postać: $$d(p, q) = \sqrt{\sum^{n}_{i=1}(q_{i}-p_{i})^2}$$

W dwuwymiarowej przestrzeni $R^{2}$ otrzymujemy:
$$ d(p, q) = \sqrt{(q_{1}-p_{1})^{2} + (q_{2}-p_{2})^{2}} $$

### Metryki Minkowskiego (uogólnienie)

Dla przestrzenie $R^{n}$

$x = (x_{1}, x_{2}, ..., x_{n}) \in R^{n}$

$y = (y_{1}, y_{2}, ..., y_{n}) \in R^{n}$

Metrykę Minkowskiego definiujemy wzorem:

$$(\sum^{n}_{i=1}|x_{i}-y_{i}|^{p})^{\frac{1}{p}}$$

Dla $p=1$ otrzymujemy <b>Metrykę Manhattan</b> (aka. matryka miasto, metryka miejska, metryka taksówkowa)

$$p=1 \Rightarrow \sum^{n}_{i=1}|x_{1}-y_{i}|$$

Dla $p=2$ otrzymujemt <b>Metrykę Euklidesową</b>

$$p=2 \Rightarrow (\sum^{n}_{i=1}(|x_{1}-y_{i})^{2}|)^\frac{1}{2} = \sqrt{\sum^{n}_{i=1}|x_{i}-y_{i}|^{2}}$$ 

Dla przestrzeni dwuwymiarowej $n=2$ otrzymujemy:

$$p=2 \land n=2 \Longrightarrow \sqrt{\sum^{2}_{i=1}|x_{i}-y_{i}|^{2}} = \sqrt{|x_{1}-y_{1}|^{2} + |x_{2}-y_{2}|^{2}}$$ 

Otrzymany wynik nazywamy wartością metryki, a po utożsamieniu metryki z odległością pomiędzy punktami (wektorami) wynik nazywamy odległością w zadanej metryce. Przestrzeń wektorową wraz ze zdefiniowaną na niej metryką nazywamy przestrzenią metryczną.

### Przykład

Niech: 

$x=(-1, 2) \in R^{2}$

$y=(3, 1) \in R^{2}$

Wtedy:

W metryce Manhattan punkty x, y oddalone są od siebie o: 
$$\sum^{n}_{i=1}|x_{1}-y_{i}| = |-1-3| + |2-1| = 4+2 = 5$$

Interpretacja geometryczna

W metryce Euklidesowej punkty x, y oddalone są od siebie o: 
$$\sqrt{\sum^{n}_{i=1}|x_{1}-y_{i}|^{2}}=\sqrt{|-1-3|^{2}+|2-1|^{2}}=\sqrt{16+4}=\sqrt{20}=4\sqrt{5}$$

### Kule w przestrzeniach metrycznych

Metrykę często przedstawia się w postaci zbioru równoodległych punktów, tzw. **kuli**.

W dwuwymiarowej przestrzeni:
- kula w metryce euklidesowej ma postać

- kula w metryce manhhatan ma postać

Do popularnych metryk należą również:
- metryka kolejowa (aka metryka centrum)
- metryka maksimum (aka metryka Czebyszewa, metryka szachowa, metryka nieskończoność)
- metryka dyskretna (aka metryka zero-jedynkowa)
