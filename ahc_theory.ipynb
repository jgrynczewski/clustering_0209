### Popularne estymatory odległość pomiędzy klastrami (linkage)

Odległość między klastrami możemy mierzyć na wiele sposób. Estymator nazywamy łaczeniem/powiązaniem (*ang. linkage*)

Do najpopularniejszych esstymatorów odległości pomiędzy klastrami należą:

#### 1. Pojedyncze łączenie

(*ang. Single-linkage*)

aka Nearest Point Algorithm

Najbliższy sąsiad.

Odległość między dwoma klastrami $C_{A}$, $C_{B}$ liczona jest jako $d_{C_{A}C_{B}}$ - minimalna odległość pomiędzy punktami klastrów.

Minusy:
- duża wrażliwość na obserwacje odstające (outliery), tzw. chaining effect

#### 2. Pełne połączenie

(*ang. Complete-linkage*)

aka Farthest Point Algorithm, Voor Hees Algorithm.


Najdalszy sąsiad

Odległość między dwoma klastrami $C_{A}$, $C_{B}$ liczona jest jako $d_{C_{A}C_{B}}$ - maksymalna odległość pomiędzy punktami klastrów.

Minusy:
- podobnie jak w single-linkage, duża wrażliwość na obserwacje odstające (outliery)

#### 3. Średnie połaczenie

(*ang. average-linkage*) 

aka UPGMA algorithm

Odległość między dwoma klastrami $C_{A}$, $C_{B}$ liczona jest jako $d_{C_{A}C_{B}}$ - uśredniona odległość pomiędzy wszystkimi punktami klastrów.

#### Kryterium Warda 

(*ang. Ward's criterion*) 

aka sum of squares criterion, Ward-linkages, incremental algorithm

Kryterium opiera się na wartości wariancji w poszczególnych klastrach. 

Mierzymy wariancję w klastrze $C_{A}$, wariancję w klastrze $C_{B}$, a następnie wariancję w klastrze powstałym z połączenia klastrów $C_{A}$ i $C_{B}$. Przyrost wariancji definuje odległość między tymi klastrami. Jeżeli przyrost jest niewielki oznacza to, że klastry są blisko siebie. Jeżeli przyrost jest duży oznacza to, że klastry są od siebie znacznie oddalone.

Inne popularne estyamtory to:
- centroid-linkage - odległość pomiędzy klastrami to odległość pomiędzy centroidami tych klastrów (aka UPGMC algorithm)
- median-linkage - odległość pomiędzy klastrami to mediana odległości pomiędzy wszystkimi punktami obu klastrów (aka WPGMC algorithm)

Dla każdego z powyższych estymatorów wynik zależy od przyjętej metryki.

### Dendogram

Wysokość dendrogramu pozwala określić stopień zbliżenia poszczególnych punktów/klastrów (im niżej łączenie, tym bliżej siebie znajdują się elementy).

Widzimy kilka sensownych cięć  (2, 3, 4, 5).

Tniemy, grupujemy i używamy poznanych miar, żeby wybrać najlepsze grupowanie.
