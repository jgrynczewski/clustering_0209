Przeprowadź analizę skupień za pomocą algorytmu ahc na zbiorze danych Eurowizji 2016, w którym za wiersze służą kraje, a za kolumny ilości punktów, które przyznały poszczególnym piosenkom - 'data/eurovision-2016.csv'

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/eurovision-2016.csv')
data
```

```python
eurovision = data.pivot(index='From country', columns='To country')['Jury Rank']
eurovision
```

```python
eurovision = eurovision.fillna(0)
eurovision
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

z = linkage(eurovision, method='complete')

dendrogram(
    z,
    labels=eurovision.index
)
plt.show()
```
